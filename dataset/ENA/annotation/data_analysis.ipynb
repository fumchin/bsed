```python
from glob import glob
import pandas as pd
import os, os.path
```

```python
recording_folder_list = glob("./Recording_*")
print(recording_folder_list)
all_recording_file_list = []
for recording_folder_path in recording_folder_list:
    recording_file_list = glob(os.path.join(recording_folder_path, "*.txt"))
    all_recording_file_list = all_recording_file_list + recording_file_list
print(all_recording_file_list)
```

```python
files = [pd.read_csv(recording_file, sep="\t") for recording_file in all_recording_file_list]
df = pd.concat(files)
print(df)
```

```python
df['Species'].value_counts()
```

```python
counts = df['Species'].value_counts()
valids = counts[counts>100].index

new_df = df[df['Species'].isin(valids)]
new_df['Species'].value_counts()
```
